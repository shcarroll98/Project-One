```python
fig, ax1 = plt.subplots()
ax1.set_ylabel('Speechiness')
ax1.boxplot(genre_data, sym = "r")
#label the x axis
ax1.set_xticklabels(genre_list)
plt.show()

#later assess the genres of each playlist, which playlists had the most songs
```
