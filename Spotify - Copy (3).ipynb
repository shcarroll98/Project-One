```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import hvplot.pandas
import pandas as pd
import numpy as np
import requests
import scipy.stats as stats

#base_url =https://api.spotify.com/v1/search
# Import API key
#from api_keys import geoapify_key
```

```python
# Store filepath in a variable
songs_path = ("Resources/spotify_songs.csv")
```

```python
# Read our data file with the Pandas library
song_data = pd.read_csv(songs_path, encoding="ISO-8859-1")
song_data
```

```python
song_data
```

```python
# Drop data with without full day/month/ year as release date.
date_filtered =song_data.loc[song_data["track_album_release_date"].str.len()>7]
date_filtered
```

```python
date_filtered.count()
```

```python
#Cleaning the file

cleaned_song_data = date_filtered
cleaned_song_data
```

```python
cleaned_song_data['track_album_release_date'] = pd.to_datetime(cleaned_song_data['track_album_release_date'])
cleaned_song_data
```

```python
cleaned_song_data = cleaned_song_data.loc[(cleaned_song_data["track_album_release_date"] >= "2010-01-01") & (cleaned_song_data["track_album_release_date"] < "2020-01-01")]

cleaned_song_data
```

```python
cleaned_track_data = cleaned_song_data.drop_duplicates(subset =['track_id'])
cleaned_track_data
```

```python
cleaned_trackid_data = cleaned_song_data.drop_duplicates(subset=['track_name', 'track_artist'], keep = False)
cleaned_trackid_data
```

```python
#create a DataFrame
spotify_df = pd.DataFrame(cleaned_trackid_data)

spotify_df
```

```python
spotify_df.groupby("playlist_genre").duration_ms.min()
#spotify_df.groupby("playlist_genre").duration_ms.max()
```

```python
#select only colunms genre and duration
genre_df = spotify_df[["playlist_genre", "duration_ms"]]
genre_df
```

```python
genre_group = genre_df.groupby("playlist_genre")
genre_group.count()
```

```python
genre_duration_df = genre_group.mean()
genre_duration_df
```

```python
plt.bar(genre_duration_df.index,genre_duration_df.duration_ms)
plt.title("Duration in MS by Genre")
plt.xlabel("Genre")
plt.ylabel("Duration Mean (ms)")


plt.show()
```

```python
bins = [0, 48000, 97000, 145000, 194000, 242000, 291000, 339000, 388000, 485000, 533000]
group_labels = ["0 to 48k", "49k to 97k", "98k to 145k", "146k to 194k", "195k to 242k", "243k to 291k", "292k to 339k", "340k to 388k", "389k to 485k", "486k to 533k"]
```

```python
pd.cut(spotify_df["duration_ms"], bins, labels=group_labels)
```

```python
genre_df["duration_group"] = pd.cut(spotify_df["duration_ms"], bins, labels=group_labels)
genre_df

```

```python
print(genre_df["duration_ms"].max())
print(genre_df["duration_ms"].min())
```
